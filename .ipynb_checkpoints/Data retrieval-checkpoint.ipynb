```python
import csv
import requests

# Set up the API call parameters
url = "https://api.rawg.io/api/games"
params = {
    "page": 1,
    "page_size": 40,
    "key": "de4c2cc9f2f0472492b8f58250c640c1"
}

fields = ["id", "name", "released", 
          "rating", 
          "platforms", "series", "genre",
          "tags","metacritic","playtime",]

# Open the CSV file for writing
with open("games.csv", mode="w", newline="", encoding="utf-8") as file:
    # Create a CSV writer object
    writer = csv.writer(file)

    # Write the header row
    writer.writerow(fields)

    # Loop through the pages of the API results
    for page in range(1, 250):
        # Set the page parameter in the API call
        params["page"] = page

        # Make the API call and get the JSON response
        response = requests.get(url, params=params)
        data = response.json()["results"]
        print(page)

        # Loop through the games on this page and write them to the CSV file
        for game in data:
            # Create variables for each field that you want to store
            series = ""
            
            params2 = {
                "key": "de4c2cc9f2f0472492b8f58250c640c1"
            }
             # Call the /games/{game_pk}/game-series endpoint to get the game series data
            game_series_url = f"https://api.rawg.io/api/games/{game['id']}/game-series"
            game_series_response = requests.get(game_series_url, params=params2)
            game_series_data = game_series_response.json()
            
            
            
            if game_series_data["results"]:
                series_ids = [game_data["id"] for game_data in game_series_data["results"]]
                series = " ".join([str(series_id) for series_id in series_ids])
                
        
            genre = ""
            if game["genres"]:
                genre = game["genres"][0]["name"]
            parent_platforms = " ".join([platform["platform"]["name"] for platform in game["parent_platforms"]]) if game["parent_platforms"] else ""
            platforms = " ".join([platform["platform"]["name"] for platform in game["platforms"]]) if game["platforms"] else ""
            tags = " ".join([tag["name"] for tag in game["tags"]]) if game["tags"] else ""
            
            
            # Write the game data to the CSV file
            writer.writerow([
                game["id"],
                game["name"],
                game["released"],
                game["rating"],
                platforms,
                series,
                genre,
                tags,
                game["metacritic"],
                game["playtime"],
            ])


```
